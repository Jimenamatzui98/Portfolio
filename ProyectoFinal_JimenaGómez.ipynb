## Introducción

La atmosfera es un factor clave para el desarrollo y el mantenimiento de la vida en la Tierra. En este sentido resulta fundamental contar con información confiable sobre los temas relacionados a la atmosfera , la cuál es útil para el análisis de sus problematicas y con la finalidad de diseñar, ejecutar acciones efectivas reflejadas en los niveles locales, nacionales y globales. No obstante, existe una relación entre los contaminates atmosféricos con la calidad del aire.
Entendemos que la calidad del aire hace referencia a las cantidades de contaminantes que se encuentran presentes en el aire. De este modo, cuando decimos que hay una buena calidad del aire, queremos decir que hay pocos contaminantes.

Cada día escuchamos que la emisión de contaminantes a la atmosfera genera daños irreversibles a las capas que cubren la Tierra y protegen al ser humano de su impacto directo de las radiaciones solares dañinas; sin embargo, hacen que cada vez tenga más efectos negativos en la salud humana debido a la alta exposición al aire contaminado.


En 2018 la Organización Mundial de la Salud (OMS) determinó que Monterrey es la ciudad más contaminada de México por la concentración alta de PM-10 y PM-2.5 en el aire. Durante dicembre del 2017 y los primeros meses del 2018 se aumento en un 15% las enfermedades respiratorias, cancerigenas, pulmonares, infartos y derrames cerebrales.
En este proyecto nos enfocaremos en el estado de Monterrey, primeramente definiremos como principal problema la mala calidad del aire y los problemas que causa en la salud; después recopilaremos los datos y los organizaremos; como tercer paso limpiaremos y transformaremos los datos de acuerdo a nuestros interesés; después analizaremos los datos con la finalidad de extraer conclusiones, predicciones que nos impusen a toma de decisiones, finalmente compartiremos nuestros resultados mostrados en un gráfico general.

Comenzamos el proces para descargar los JSON de la API, convertirlos a CSV y a DATAFRAME

```python
from pathlib import Path
import json
#3243971 dataset from sempember 4th 2018, 5000 per page, 649 pages.
for page in range(194,269):
  #saving the complete json file per page
    json_file='page_%i_sinaica_mediciones.json'%page
    json_file+'/mediciones_sinaica_json/'+'json_file'
    #checkig if the file already exist for not having it double
    if not Path(json_file).is_file():
      #requesting to the API using 5000 registers per page, for every page
      data_api= client.makeCall('sinaica', {'pageSize':5000,'page':page})
      #Writing the file
      with open(json_file, 'w') as outfile:
       json.dump(data_api, outfile)
```

```python
def parse_mediciones_json(json_file):
  with open(json_file, 'r') as aux:
    results= json.load(aux)['results']

  pre_data= []

  for r in results:
    aux= pd.DataFrame.from_dict(r, orient='index').T
    pre_data.append(aux)

  if len(pre_data) > 0:
    pre_data = pd.concat(pre_data, ignore_index= True)
    return pre_data

  else:
    print('NO DATA FOR', json_file)
    return
```

```python
from joblib import Parallel, delayed

name= '/content/page_%i_sinaica_mediciones.json'
!mkdir mediciones_sinaica_csv
!mkdir data

contador= 1

for i, j in zip([194,230], [230,269]):
  print('Chunk', contador)
  aux_list= [name %n for n in range(i, j)]
  sinaica_mediciones= Parallel(n_jobs=-1, verbose= 8)(delayed (parse_mediciones_json)(f) for f in aux_list)

  sinaica_mediciones= pd.concat(sinaica_mediciones, ignore_index=True)
  print(sinaica_mediciones.shape)

  sinaica_mediciones.to_csv('/content/mediciones_sinaica_csv/sinaica_mediciones_%i_%i.csv'%(i,j), index= False)

  contador+=1

from glob import glob

all_data_mediciones= []

for file in glob ('/content/mediciones_sinaica_csv/*.csv'):
  aux= pd.read_csv(file, usecols= ['estacionesid', 'fecha', 'hora', 'city','state', 'parametro', 'valororig', 'validoorig'])
  all_data_mediciones.append(aux)

all_data_mediciones= pd.concat(all_data_mediciones)
all_data_mediciones.to_csv('./data/mediciones_todas_estanciones.csv', index=False)
```

Al parecer solo hubo 8 entradas con estos valores, en una misma fecha (2018-02-11) entre las 2 am- 9 am, lo que pudo haber sido una falla del sensor, por lo que al ser una cantidad de muestra muy pequeña, la tomaremos como outlier, y sera removida con el IQR.


```python
#Remover outliers del dataframe de monoxido de carbono en las 11 estaciones de monterrey y volver a graficar el boxplot
Q1= CO['valororig'].quantile(0.25)
Q2= CO['valororig'].quantile(0.75)
IQR= Q2-Q1
LI = Q1-(1.5*IQR)
LS = Q2+(1.5*IQR)
outliers_CO= CO[(CO['valororig']>= LI) & (CO['valororig']<= LS)]
plt.title('Boxplot CO en Monterrey (2018) sin outliers')
plt.ylabel('Valor (ppm)')
B2= sns.boxplot(outliers_CO['valororig'])
plt.show()
```
