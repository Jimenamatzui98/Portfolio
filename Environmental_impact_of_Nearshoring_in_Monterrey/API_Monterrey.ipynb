```python
from datosgobmx import client
```

```python
url= 'https://api.datos.gob.mx/v1/calidadAire'
```

```python
import requests
response=requests.get(url)
dicc= response.json()
dicc.keys()

meas=[]
for r in dicc['results']:
  aux= pd.DataFrame.from_dict(r, orient='index').T
  meas.append(aux)

meas= pd.concat(meas, ignore_index=True)
meas.head(10)
```

```python
#crear un dataframe para ver los parámetros (clasificación de los componentes del aire)
#dióxido de azufre (SO2), ozono (O3), monóxido de carbono (CO) y dióxido de nitrógeno (NO2),
# reportadas en partículas por millón (ppm); temperatura (TMP), reportada en grados centígrados;
#partículas menores a 2.5 micrómetros (PM2.5) y partículas menores a 10 micrómetros (PM10),
#reportadas en microgramos por metro cúbico (μg/m3).
climate_params=[]
for x in request['results']:
  aux= pd.DataFrame.from_dict(x, orient='index').T
  climate_params.append(aux)

climate_params
climate_params= pd.concat(climate_params, ignore_index=True)
climate_params
```

```python
#To locate the meteorological stations in mexico
request_stations= client.makeCall('sinaica-estaciones', {'pageSize':200})
request_stations.keys()
```

```python
stations=[]
for s in request_stations['results']:
  aux= pd.DataFrame.from_dict(s, orient='index').T
  stations.append(aux)

stations= pd.concat(stations, ignore_index=True)
stations.head()
```

```python
import folium
#Center the map
centro_lat, centro_lon = 22.396092, -101.731430
#map generation
Mexico_airquality_stations= folium.Map(location=[centro_lat, centro_lon], zoom_start=5, tiles='cartodb positron')
#fill the data in the map
for i, row in stations.iterrows():
  popup_text= f'<b> Nombre: </b> {row.nombre} <br> <b> Latitud: </b> {row.lat:.5f} <br> <b> Longitud: </b> {row.long: .5F} <br>'
  folium.CircleMarker(location=[row.lat, row.long], radius=3, tooltip=popup_text, fill=True, fill_opacity= 0.4).add_to(Mexico_airquality_stations)

Mexico_airquality_stations
```

```python
#For the measurements in each station
measurements_request= client.makeCall('sinaica', {'pageSize':5000})
measurements_request.keys()
```

```python
measurements=[]
for m in measurements_request['results']:
  aux= pd.DataFrame.from_dict(m, orient='index').T
  measurements.append(aux)

measurements= pd.concat(measurements, ignore_index=True)
measurements.sort_values('date', ascending=False)
```
